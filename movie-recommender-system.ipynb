# Preprocessing

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
import json

def convert(obj):
    L = []
    for i in json.loads(obj):
        L.append(i['name'])
    return L    

movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)


print(movies['genres'][0])
print(movies['keywords'][0])
```

```python
def convert_cast(obj):
    L = []
    counter = 0
    for i in json.loads(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies["cast"] = movies["cast"].apply(convert_cast)
print(movies['cast'][1])
```

```python
# I need those values where job == director
movies["crew"][0]

def fetch_director(obj):
    L = []
    for i in json.loads(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L

movies["crew"] = movies["crew"].apply(fetch_director)
print(movies['crew'][0])
```

```python
movies.head(1)
```

```python
movies["overview"][0]

movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies["genre"] = movies["genres"].apply(lambda x: [i.replace(" ", "") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x: [i.replace(" ", "") for i in x])
movies["cast"] = movies["cast"].apply(lambda x: [i.replace(" ", "") for i in x])
movies["crew"] = movies["crew"].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x).lower())
```

```python
new_df['tags'][0]
```

```python
import nltk
```

```python
# loved, love, loving --> love
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem("loving")
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors.shape
```

```python
vectors[0]
```

```python
# Lets find the similarity between vectors
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
def recommend(movie, top_n=5):
    """
    Return a list of `top_n` movie titles most similar to `movie`.
    """
    print("="*60)
    print(f"üé¨ Looking for recommendations for: '{movie}'")
    print("="*60)
    
    # STEP 1: Find the movie in our database
    print("\nüìç STEP 1: Finding the movie in our database...")
    matches = new_df[new_df['title'] == movie].index
    print(f"   Matches found: {list(matches)}")
    
    if len(matches) == 0:
        print("   ‚ùå Movie not found in database!")
        return []
    
    movie_index = matches[0]
    print(f"   ‚úÖ Found '{movie}' at index (row number): {movie_index}")
    
    # STEP 2: Get similarity scores for this movie
    print("\nüìä STEP 2: Getting similarity scores...")
    distances = similarity[movie_index]
    print(f"   Total movies in database: {len(distances)}")
    print(f"   Similarity scores (first 10): {distances[:10]}")
    print(f"   Score at index {movie_index} (the movie itself): {distances[movie_index]}")
    
    # STEP 3: Pair each movie index with its similarity score
    print("\nüî¢ STEP 3: Pairing movie indices with scores...")
    enumerated = list(enumerate(distances))
    print(f"   First 5 pairs (index, score):")
    for i in range(min(5, len(enumerated))):
        idx, score = enumerated[i]
        movie_title = new_df.iloc[idx].title
        print(f"      Index {idx}: '{movie_title}' ‚Üí similarity = {score:.4f}")
    
    # STEP 4: Sort by similarity score (highest first)
    print("\nüìà STEP 4: Sorting by similarity (highest first)...")
    scored = sorted(
        enumerate(distances),
        key=lambda x: x[1],
        reverse=True
    )
    print(f"   Top 5 most similar (including the movie itself):")
    for i in range(min(5, len(scored))):
        idx, score = scored[i]
        movie_title = new_df.iloc[idx].title
        print(f"      #{i+1}: '{movie_title}' (index {idx}) ‚Üí similarity = {score:.4f}")
    
    # STEP 5: Skip the first one (the movie itself) and take top_n
    print(f"\n‚úÇÔ∏è STEP 5: Removing the movie itself and taking top {top_n}...")
    print(f"   Skipping: {new_df.iloc[scored[0][0]].title} (the movie itself)")
    top_matches = scored[1: top_n + 1]
    print(f"   Selected {len(top_matches)} recommendations:")
    for i, (idx, score) in enumerate(top_matches):
        movie_title = new_df.iloc[idx].title
        print(f"      #{i+1}: '{movie_title}' ‚Üí similarity = {score:.4f}")
    
    # STEP 6: Extract just the titles
    print("\nüìù STEP 6: Extracting movie titles...")
    recommendations = [new_df.iloc[i[0]].title for i in top_matches]
    print(f"   Final recommendations: {recommendations}")
    
    print("\n" + "="*60)
    print(f"‚úÖ DONE! Returning {len(recommendations)} recommendations")
    print("="*60 + "\n")
    
    return recommendations
```
